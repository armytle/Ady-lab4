```python
# import libraries
import pandas as pd
from sqlalchemy import create_engine
```

```python
# load messages dataset
df =  pd.read_csv('augmented_data.csv')
df.head()
```

```python
df = df[df.columns[:40]]
df.head()
```

```python
for col in df:
  print(col,' : ',df[col].unique())
```

```python
column_list = df.columns.tolist()
print(column_list)
print(len(column_list))
```

```python
# check number of duplicates
df.duplicated().sum()
```

```python
# drop duplicates
df= df.drop_duplicates()
```

```python
df = df.drop('child_alone',axis = 1)
```

```python
df['related'] = df['related'].apply(lambda x: 1 if x==2 else x)
```

```python
new_order = [
    'offer', 'shops', 'tools', 'fire', 'hospitals', 'missing_people',
    'aid_centers', 'clothing', 'security', 'cold', 'electricity',
    'money', 'search_and_rescue', 'military', 'refugees', 'other_infrastructure',
    'death', 'transport', 'medical_products', 'buildings', 'other_weather',
    'water', 'infrastructure_related', 'medical_help', 'floods', 'shelter',
    'storm', 'earthquake', 'food', 'other_aid', 'request', 'direct_report',
    'related', 'weather_related', 'aid_related'
]

# Lấy danh sách các cột hiện tại
current_columns = df.columns.tolist()

# Giữ nguyên các cột từ 0 đến 3
fixed_columns = current_columns[:4]

# Thêm các cột từ thứ 4 theo thứ tự mới
new_columns_order = fixed_columns + new_order

# Đặt lại thứ tự các cột trong DataFrame
df = df[new_columns_order]

# Kiểm tra lại thứ tự các cột
print(df.columns)
```

```python
engine = create_engine('sqlite:///DisasterResponse.db')
df.to_sql('DisasterResponse', engine, index=False, if_exists='replace')
```
